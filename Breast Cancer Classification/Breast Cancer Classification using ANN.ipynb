```python
import numpy as np
import pandas as pd
import sklearn.datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset.DESCR)
```

```python
data = pd.DataFrame(data=breast_cancer_dataset['data'], columns=breast_cancer_dataset['feature_names'])
```

```python
data['target'] = breast_cancer_dataset['target']
```

```python
X = data.drop(columns='target', axis=1)
y = data['target']
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.losses import BinaryCrossentropy, SparseCategoricalCrossentropy
from tensorflow.keras.optimizers import Adam
```

```python
tf.random.set_seed(3)
```

```python
model = Sequential([
    tf.keras.Input(shape=(30,)),
    Dense(units=16, activation='relu'),
    Dense(units=8, activation='relu'),
    Dense(units=1, activation='sigmoid'),    
])
```

```python
model.compile(loss=BinaryCrossentropy(), optimizer=Adam(0.001), metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(X_train_std,y_train,validation_split=0.1, epochs=20)
```

```python
y_pred = model.predict(X_test_std)
```

```python
y_pred
```

```python
y_pred = (y_pred>=0.5)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
cr = classification_report(y_pred=y_pred, y_true=y_test)
print(cr)
```

```python
print(accuracy_score(y_pred=y_pred,y_true=y_test))
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')

plt.legend(['training data', 'validation data'])
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')

plt.legend(['training data', 'validation data'], loc='upper right')
```
